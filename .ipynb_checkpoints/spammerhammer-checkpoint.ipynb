```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import re
from wordcloud import WordCloud, STOPWORDS
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn import metrics, svm
from sklearn.naive_bayes import BernoulliNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
```

# Adatok betöltése

```python
def load_uci_data_csv(file_path):
    # https://archive.ics.uci.edu/ml/datasets/YouTube+Spam+Collection
    result = []
    full_text = ""
    with open(file_path, encoding="utf-8") as csv_file:
        for i, row in enumerate(csv.reader(csv_file, delimiter=',')):
            if i == 0:
                continue

            result.append([row[3], float(row[4])])
            
            if i == 1:
                full_text += row[3]
            else:
                full_text += " " + row[3]
            
    return result, full_text

# gyönyörű megoldás :D
data1, full_text1 = load_uci_data_csv("./Youtube01-Psy.csv")
data2, full_text2 = load_uci_data_csv("./Youtube02-KatyPerry.csv")
data3, full_text3 = load_uci_data_csv("./Youtube03-LMFAO.csv")
data4, full_text4 = load_uci_data_csv("./Youtube04-Eminem.csv")
data5, full_text5 = load_uci_data_csv("./Youtube05-Shakira.csv")

full_text = full_text1 + " " + full_text2 + " " + full_text3 + " " + full_text4 + " " + full_text5
```

# 1. Analízis
## 1.1. Szófelhő mert azt mindenki szereti

```python
def generate_word_cloud(data):
    wc = WordCloud(background_color='white',
                   stopwords=STOPWORDS,
                   max_words=300,
                   max_font_size=40, 
                   random_state=42).generate(data)
    
    plt.figure(figsize=(7, 15))
    plt.imshow(wc)
    plt.show()
    
generate_word_cloud(full_text)
```

## 1.2. Lehetne egyéb dolgokat vizsgálni...

Meg kellene ragadni azokat a tulajdonságokat, amik a spamot spammá teszik... Én most csak a szavak számára és a szavak frekvenciájára fogok hagyatkozni.

# 2. Tanító és teszthalmaz létrehozása

```python
def append_data(X, y, data_set):
    for row in data_set:
        X.append(row[0])
        y.append(row[1])


train_X = []
train_y = []
test_X = []
test_y = []

append_data(train_X, train_y, data1)
append_data(train_X, train_y, data2)
append_data(train_X, train_y, data3)
append_data(train_X, train_y, data4)

append_data(test_X, test_y, data5)
```

# 3. Modell: Random találgatás

```python
def test_random_guess(state=42):
    np.random.seed(state)
    
    predictions = np.random.randint(2, size=len(test_y))
    
    print(metrics.classification_report(test_y, predictions))
    
test_random_guess()
```

# 4. Modell: Lineáris SGD

```python
def train_sgd(loss="hinge"):
    model_pipeline = Pipeline([('cnts', CountVectorizer()),
                               ('tfidf', TfidfTransformer()),
                               ('clf', SGDClassifier(loss=loss, random_state=42))])
    
    model_pipeline.fit(train_X, train_y)
    
    return model_pipeline

def test_sgd(loss="hinge"):
    model = train_sgd(loss=loss)
    predictions = model.predict(test_X)
    
    return predictions
    
print(metrics.classification_report(test_y, test_sgd("hinge")))
```

# 5. Modell: Logisztikus SGD

```python
print(metrics.classification_report(test_y, test_sgd(loss="log")))
```

# 6. Modell: SVM

```python
def train_svm(kernel="linear"):
    model_pipeline = Pipeline([('cnts', CountVectorizer()),
                               ('tfidf', TfidfTransformer()),
                               ('clf', svm.SVC(kernel=kernel, random_state=42))])
    
    model_pipeline.fit(train_X, train_y)
    
    return model_pipeline

def test_svm(kernel="linear"):
    model = train_svm(kernel=kernel)
    predictions = model.predict(test_X)
    
    return predictions

print(metrics.classification_report(test_y, test_svm("linear")))
```

# 7. Modell: Naív Bayes

```python
def train_naive_bayes():
    model_pipeline = Pipeline([('cnts', CountVectorizer()),
                               ('tfidf', TfidfTransformer()),
                               ('clf', BernoulliNB())])
    
    model_pipeline.fit(train_X, train_y)
    
    return model_pipeline

def test_naive_bayes():
    model = train_naive_bayes()
    predictions = model.predict(test_X)
    
    return predictions
    
print(metrics.classification_report(test_y, test_naive_bayes()))
```

# 8. Modell: Random forest (Csak az érdekesség kedvéért)

```python
def train_random_forest():
    model_pipeline = Pipeline([('cnts', CountVectorizer()),
                               ('tfidf', TfidfTransformer()),
                               ('clf', RandomForestClassifier(random_state=42))])
    
    model_pipeline.fit(train_X, train_y)
    
    return model_pipeline

def test_random_forest():
    model = train_random_forest()
    predictions = model.predict(test_X)
    
    return predictions
    
print(metrics.classification_report(test_y, test_random_forest()))
```

# 9. Többrétegű perceptron (MLP)

```python
def train_mlp():
    model_pipeline = Pipeline([('cnts', CountVectorizer()),
                               ('tfidf', TfidfTransformer()),
                               ('clf', MLPClassifier(hidden_layer_sizes=100, activation="relu", solver="adam",
                                                    random_state=42, early_stopping=True, max_iter=1000))])
    
    model_pipeline.fit(train_X, train_y)
    
    return model_pipeline

def test_mlp():
    model = train_mlp()
    predictions = model.predict(test_X)
    
    return predictions
    
print(metrics.classification_report(test_y, test_mlp()))
```

# Eredmények összehasonlítása

```python
def plot_roc_curves(gt, predictions, names):
    plt.figure(figsize=(8, 8))
    plt.title('ROC')
    
    plt.plot([0,1],[0,1],'--', label="Random")
    
    leaderboard = []
    
    for i in range(len(predictions)):
        fpr, tpr, thresholds = metrics.roc_curve(gt, predictions[i])
        roc_auc = metrics.auc(fpr, tpr)
        leaderboard.append([names[i], roc_auc])
#         print("{}: {}".format(names[i], roc_auc))
    
        plt.plot(fpr, tpr, label=names[i])
           
    plt.legend(loc='lower right')
    plt.xlim([-0.1,1.0])
    plt.ylim([-0.1,1.0])
    plt.ylabel('True positive ráta')
    plt.xlabel('False positive ráta')
    plt.show()
    
    leaderboard.sort(key=lambda x: x[1], reverse=True)
    
    print("Modellek végső rangsora:")
    print("------------------------")
    for i, elem in enumerate(leaderboard):
        print("{}. {}: ROC AUC = {}".format(i + 1, elem[0], elem[1]))
    
def plot_final_roc():
    pred_sgd_linear = test_sgd(loss="hinge")
    pred_sgd_log = test_sgd(loss="log")
    pred_svm = test_svm(kernel="linear")
    pred_nb = test_naive_bayes()
    pred_rf = test_random_forest()
    pred_mlp = test_mlp()
    
    plot_roc_curves(test_y, [pred_sgd_linear,
                             pred_sgd_log,
                             pred_svm,
                             pred_nb, 
                             pred_mlp, 
                             pred_rf], 
                    ["SGD - Lineáris", 
                     "SGD - Logisztikus",
                     "SVM",
                     "Naív Bayes", 
                     "MLP", 
                     "Random Forest"])
    
plot_final_roc()    
```
